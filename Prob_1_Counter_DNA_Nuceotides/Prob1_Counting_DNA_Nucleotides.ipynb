# Problem 1: Counting DNA Nucleotides 

### Problem
A **string** is simply an ordered collection of symbols selected from some alphabet and formed into a word; the length of a string is the number of symbols that it contains.

An example of a length 21 DNA string (whose alphabet contains the symbols 'A', 'C', 'G', and 'T') is "ATGCTTCAGAAAGGTCTTACG."

**Given:** 
A DNA string s of length at most 1000 nt.

**Return:** 
Four integers (separated by spaces) counting the respective number of times that the symbols 'A', 'C', 'G', and 'T' occur in s. 

**Sample Dataset**
AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC

**Sample Output**
20 12 17 21

```python
s = "GCCGCGGACCGAACATCCGACAAGTCATTGTATTGCGATGGGGAGAACCACGGGTTGCATCCTCCGGGCAACCGAGTGTGGCTTCAGCCGTAACACCTCTGAAAACCTCACGAGTGACGTGATATTGCTCCTTTTTACTTGGGGGCGGTCCCGCGCGGTAGGTGTCGGAGAAAGATGAACGCCTGATTACATAAGGTCCCTTGGATTTGGGGGTACACAGAGCTGGAAAGCGGTGAACGACTCTTGATTGGACACATCCAGAAATTTAAACATTAGCCTTAACCTGGTTCGCAAATTGTAGTCAGATGGTAGATTTCCTATTGAATATCACCGTTGGTCAGGATGCTTGTTTCTGTAATCACGCTTGAAGCTCCCCAGTTACCGTCTAGCCTGATAAAGTAACTCCTAGGACCTGCGAAGGTTTATCCCCGCAGTGGTTTACCAAGCCATTTTTCATTCTCATAAGACTGGGTGCAACATACGTCCCCAACTGGGATTTTTGGGCCTAACTTTACCGCCTGGACCACCAGACATTGTTCCTCAGCGTCGCACAGCCAACCTTTGGCCCCAATAAGGAAAGCAGTCTGTTGGGCATTCCACTCTACACATGTCTGGGAGAAGAAGGGGCTGCCCCGTAGCTTCCGTAACGGGTCGTACGTTGAAGTAAAGTCTTACGGCTAGATTGGACAGAACTAACTCTTAGGTAGAGCGGGACGGCACCAACATCTGACCCCCTCATTTGCCGCTTAGTACGAGGAATGCGTACCCACTGTAGTATTGGTGCCGCTGCTCCAACTGGCAGAGGATGTATCCGAAGAGCTAGCATACGGCTTATATAGAGTT"
a_counter = 0
c_counter = 0 
g_counter = 0 
t_counter = 0

for letter in s: 
    if letter == "A": 
        a_counter+=1
    elif letter == "C": 
        c_counter+=1
    elif letter == "G": 
        g_counter+=1
    else:
        t_counter+=1

print(a_counter, c_counter, g_counter, t_counter)
```
